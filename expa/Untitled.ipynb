```python
f = open("expA_data.txt", "r")
txt = f.readline()
# txt是否为空可以作为判断文件是否到了末尾
data_list=[]
while txt:
    #data_list.append(txt.split(" ",1))
    data_list.append(txt.split())
    txt = f.readline()
f.close()
    

```

```python
print(data_list)
```

```python
data_dict={}
for i in range(len(data_list)):
    data_dict[data_list[i][1]]=data_list[i][0]
print(data_dict)
```

```python
number_list=[]
for j in data_dict.keys():
    number_list.append(int(j))
```

```python
def partition(arr, low: int, high: int):
    pivot, j = arr[low], low
    for i in range(low+1, high + 1):
        if arr[i] <= pivot:
            j += 1
            arr[j], arr[i] = arr[i], arr[j]
    arr[low], arr[j] = arr[j], arr[low]
    return j 

def quick_sort_between(arr, low: int, high: int):
    if high-low <= 1: # 递归结束条件
        return

    m = partition(arr, low, high)  # arr[m] 作为划分标准
    quick_sort_between(arr, low, m - 1)
    quick_sort_between(arr, m + 1, high)

def quick_sort(arr):
    """
    快速排序(in-place)
    :param arr: 待排序的List
    :return: 快速排序是就地排序(in-place)
    """
    quick_sort_between(arr,0, len(arr) - 1)

# 测试数据

print("原始数据：", number_list)
quick_sort(number_list)
quick_sort(number_list)
print("快速排序结果：",number_list)

```
