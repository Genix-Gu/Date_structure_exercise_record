```python
import random
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def order_sequential_search(alist,item):
    pos=0
    found =False
    stop=False
    while pos <len(alist) and not found and not stop:
        if alist[pos] == item:
            stop=True
        else:
            pos=pos+1
    return found
def binary_search(alist,item):
    first=0
    last=len(alist)-1
    found=False
    while first<=last and not found:
        mid_point=(first+last)//2
        if alist[mid_point] ==item:
            found=True
        else:
            if item<alist[mid_point]:
                last=mid_point-1
            else:
                first=mid_point+1
        return found
```

```python
order_time=[]
binary_time=[]
for i in range(1000):
    list_len=i*50
    random_list=[]
    for j in range(list_len):
        random_list.append(random.randint(0,list_len))
    random_list.sort()
    item=random.randint(0,list_len)
    order_start_time=time.time_ns()
    find_order=order_sequential_search(random_list,item)
    order_end_time=time.time_ns()
    order_time.append(order_end_time-order_start_time)
    binary_start_time=time.time_ns()
    find_binary=binary_search(random_list,item)
    binary_end_time=time.time_ns()
    binary_time.append(binary_end_time-binary_start_time)
plt.plot(order_time,label="order")
plt.plot(binary_time,label="binary")
plt.legend(loc='upper left')
# print(order_time)
# print(binary_time)
```
