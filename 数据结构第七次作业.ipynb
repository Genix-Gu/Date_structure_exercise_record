```python
import random
def two_dir_bubble(queue,show=False):
    dir=True
    forward_num=0
    backward_num=0
    
    for i in range(len(queue)):
        #print(queue)
        changed=False
        if(dir):
            forward_num+=1
            for j in range(backward_num,len(queue)-forward_num):
                if(queue[j]>queue[j+1]):
                    queue[j],queue[j+1]=queue[j+1],queue[j]
                    changed=True
        else:
            backward_num+=1
            for j in range(len(queue)-forward_num,backward_num-1,-1):
                if(queue[j]<queue[j-1]):
                    queue[j],queue[j-1]=queue[j-1],queue[j]
                    changed=True
        dir=not dir
        if not changed:
            break
        if show:
            print(queue) 
    #print("total epoch is:",forward_num+backward_num)
    #print("new_queue:",queue)
    return forward_num+backward_num
```

```python
def bubbleSort(arr):
    n = len(arr)
 
    # 遍历所有数组元素
    epoch_num=0
    for i in range(n):
        changed=False
        epoch_num+=1
        # Last i elements are already in place
        for j in range(0, n-i-1):
 
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                changed=True
        if not changed:
            break
    #print("total_epoch is",epoch_num)
    return epoch_num
```

```python
def generate_rand_queue(length):
    rand_queue=[]
    for i in range(length):
        rand_queue.append(random.randint(0,length))
    return rand_queue
```

```python
two_dir_bubble(generate_rand_queue(20),show=True)
bubbleSort(generate_rand_queue(20))
```

```python
import copy
import matplotlib.pyplot as plt
%matplotlib inline
two_dir_epoch=[]
one_dir_epoch=[]
for i in range(10):
    for j in range(10):
        rand_list=generate_rand_queue((i+1)*10)
        two_dir_list=copy.deepcopy(rand_list)
        one_dir_list=copy.deepcopy(rand_list)
        two_dir_epoch.append(two_dir_bubble(two_dir_list)/len(rand_list))
        one_dir_epoch.append(bubbleSort(one_dir_list)/len(rand_list))
plt.title("The ratio of sort epoch and the queue length")
plt.plot(two_dir_epoch,c="red",label="two_direction")
plt.plot(one_dir_epoch,c="blue",label="one_direction")
plt.legend()
```
