```python
class Stack:
    def __init__(self):
        self.stack = []

    def __len__(self):
        return len(self.stack)

    def push(self, data):
        self.stack.append(data)

    def pop(self):
        return self.stack.pop()

    def is_empty(self):
        if not self.stack:
            return True
        else:
            return False

    def get_top(self):
        return self.stack[-1]


def mid2back(string):

    expression = []  # 存放表达式
    sign_stack = Stack()  # 存放运算符和开括号
    sign_table = {
        '*': 2,
        '/': 2,
        '%': 2,
        '+': 1,
        '-': 1,
        '(': 0,
        ')': 0,
    }
    print("==========================================")
    for ch in string:
        if ch == '(':
            sign_stack.push(ch)
        elif ch == ')':
            while sign_stack.get_top() != '(':
                expression.append(sign_stack.pop())
            sign_stack.pop()
        elif ch == ' ':
            continue
        elif ch.isupper():
            expression.append(ch)
        else:
            while not sign_stack.is_empty() and sign_table[sign_stack.get_top()] >= sign_table[ch]:
                expression.append(sign_stack.pop())
            sign_stack.push(ch)
        print(string,sign_stack.stack)
        print(expression)
    while not sign_stack.is_empty():
        if sign_stack.get_top()=="()":
            print("don't match")
            break
        
        expression.append(sign_stack.pop())

    return "".join(expression)


if __name__ == '__main__':
    expressions = [
        '(A + B) * (C + D) * (E + F)',
        'A + ((B + C) * (D + E))',
        'A * B * C * D + E + F',
    ]

    for expression in expressions:
        print(expression, ': ', mid2back(expression))
```
